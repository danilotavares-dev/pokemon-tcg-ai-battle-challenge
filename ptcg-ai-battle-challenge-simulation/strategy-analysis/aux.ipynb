```python
import pandas as pd
```

```python
df = pd.read_csv('../utils/Card_Data.csv')
```

```python
pokemon_id = 35

all_attack_costs = df.loc[df['Card ID'] == pokemon_id, 'Cost']

int_costs = []

for cost_string in all_attack_costs:
    if pd.isna(cost_string):
        int_costs.append(0)
    else:
        total_energy = cost_string.count('{') + cost_string.count('●')
        int_costs.append(total_energy)

biggest_attack_cost = max(int_costs)

print(all_attack_costs)
biggest_attack_cost
```
